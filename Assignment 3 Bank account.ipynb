```python
class BankAccount:
    def __init__(self,name,accountNo,accountType,bal):
        self.name = name
        self.accountNo = accountNo
        self.accountType = accountType
        self.bal = bal
        
    def deposit(self,amount):
        self.bal += amount
        print(f"Deposit of {amount} is successful. New balance is {self.bal}")
    
    def withdraw(self,amount):
        if self.bal >= amount:
            self.bal -= amount
            print(f"Withdrawal of {amount} is successful. New balance is {self.bal}")
        
        else :
            print("Insufficient Balance")
            
    def display(self):
        #rint("Name:" {self.name})
        print(f"Name: {self.name}")
        print(f"Account Number : {self.accountNo}")
        print(f"Account Type : {self.accountType}") 
        print(f"Balance : {self.bal}")
        
        
while True :
    print("1 to Open new account")
    print("2 to deposit")
    print("3 to withdraw")
    print("4 Account Details")
    print("5 to Quit")
        
    choice = input()
    if choice == "1":
        name = input("Enter your name : ")
        accountNo = input("Enter your account number : ")
        accountType = input("Enter your account type :")
        bal =  float(input("Enter your balance : "))
        
        account = BankAccount(name,accountNo,accountType,bal)
        print("Account Created")
        
        
    elif choice == "2":
        if "account" in locals():
            amount = float(input("Enter amount to deposit:"))
            account.deposit(amount)
            print("Account created ")
            
    elif choice == "3":
        if "account" in locals():
            amount = float(input("Enter amount to withdraw:"))
            account.withdraw(amount)
        else:
            print("Create an account first:")
            
    elif choice == "4":
        if "account" in locals():
            account.display()
            
        else:
            print("Create an account first:")
    
    elif choice == "5":
        break
    
    else :
        print("Invalid choice")
            
```
